```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
import time
n_values = list(range(10000))
dist_values = []
for e in range(3):
    print(time.strftime('%H:%M'))
    print('Expirement: {}'.format(e))
    dist_values.append([])
    for n in n_values:
        initial_point = [0,0]
        for i in range(n):
            r = np.random.choice(['left', 'right', 'up', 'down'])
            if r == 'left':
                initial_point[0] -= 1
            elif r == 'right':
                initial_point[0] += 1
            elif r == 'up':
                initial_point[1] += 1
            else:
                initial_point[1] -= 1
        dist_values[e].append(np.round(np.sqrt(initial_point[0] ** 2 + initial_point[1] ** 2), 2))
```

```python
fig, ax = plt.subplots(figsize=(20,12))
# ax.plot(n_values, dist_values[0], color='tab:blue', label='experiment 1');
sns.regplot(n_values, dist_values[0], order=3, color='#003f5c', ci=95, fit_reg=True, scatter_kws={'s': 2, 'alpha':0.5}, 
            line_kws={'alpha':0.5, 'label':'regression line', 'linewidth': 2}, ax=ax)
sns.regplot(n_values, 0.9*np.sqrt(n_values), order=3, color='#ff7c43', ci=95, 
            scatter=False, line_kws={'alpha':1.0}, ax=ax, label='0.9 x n')
plt.legend()
plot_conf(ax, xlbl='n', ylbl='Distance')
plt.savefig('p-3.png', dpi=300)
```
